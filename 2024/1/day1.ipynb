# Part 2

```python
data = np.genfromtxt("input", dtype=int)
list_1 = data[:, 0]
list_2 = data[:, 1]

sorted_unique_1, counts_1 = np.unique(list_1, return_counts=True)
sorted_unique_2, counts_2 = np.unique(list_2, return_counts=True)
```

```python
print(sorted_unique_1)
print(counts_1)
print(sorted_unique_2)
print(counts_2)
```

```python
sum = 0
for num in list_1:
    if num in sorted_unique_2:
        index = np.where(sorted_unique_2 == num)[0][0]
        sum += num * counts_2[index]
        # print(f"{num=} {counts_2[index]=} {sum=}")
print(sum)
```
